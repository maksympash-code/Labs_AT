## Завдання 1. 

__Знайдіть усі незвідні системи твірних групи $S_4$.__

```python
G = SymmetricGroup(4)

print(f"Група: {G}")
print(f"Порядок групи: {G.order()}")

generators = G.gens()
print("Можлива система твірних групи S_4:")
for gen in generators:
    print(gen)

print("Перевірка, чи генерують твірні групу:")
print(G == G.subgroup(generators))
```

## Завдання 2. 

__Знайдіть усі підгрупи груп $S_4, A_4, S_5, A_5$.__

```python
# Функція для знаходження та виведення всіх підгруп
def list_subgroups(group):
    print(f"Підгрупи групи {group}:")
    subgroups = group.subgroups()
    for i, H in enumerate(subgroups):
        print(f"{i + 1}: Порядок підгрупи: {H.order()}, Елементи: {H.list()}")
    print(f"Загальна кількість підгруп: {len(subgroups)}\n")

S4 = SymmetricGroup(4)
A4 = AlternatingGroup(4)
S5 = SymmetricGroup(5)
A5 = AlternatingGroup(5)

list_subgroups(S4)
list_subgroups(A4)
list_subgroups(S5)
list_subgroups(A5)
```

## Завдання 3. 

__Задайте дві випадкові підстановки з групи $S_n, \, (1 \le n \le 100)$. Перевірте чи є вони парними. Знайдіть порядок групи, ними породженої. Чи буде ця група дорівнювати $A_n$? $S_n$?__

```python
import random

def generate_random_permutations(n):
    G = SymmetricGroup(n)
    return G.random_element(), G.random_element()

def check_parity(perm):
    return perm.sign() == 1

def generated_group_order(perms):
    G = SymmetricGroup(perms[0].parent().degree())
    H = G.subgroup(perms)
    return H.order(), H

def is_equal_to_A_or_S(group, n):
    An = AlternatingGroup(n)
    Sn = SymmetricGroup(n)
    return group.is_isomorphic(An), group.is_isomorphic(Sn)

n = 5

perm1, perm2 = generate_random_permutations(n)
print(f"Перша підстановка: {perm1}")
print(f"Друга підстановка: {perm2}")

print(f"Перша підстановка парна: {check_parity(perm1)}")
print(f"Друга підстановка парна: {check_parity(perm2)}")

order, generated_group = generated_group_order([perm1, perm2])
print(f"Порядок породженої групи: {order}")

is_An, is_Sn = is_equal_to_A_or_S(generated_group, n)
print(f"Група дорівнює A_{n}: {is_An}")
print(f"Група дорівнює S_{n}: {is_Sn}")
```
