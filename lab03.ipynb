__Задайте групи $D_{10}$, $Z_{16}$, $A_5$, $SL(2, \mathbb{Z}_3)$*. Для кожної з них виведіть таблицю Келі, знайдіть порядок групи, перевірте чи група абелева. Перевірте чи будуть ізоморфними групи $SL(2, Z_3)$ та $D_{12}$.__

```python
# D₁₀ - діедральна група порядку 10
G = DihedralGroup(5)
print("Таблиця Келі для D₁₀:")
print(G.cayley_table())
print("Порядок групи D₁₀:", G.order())
print("Чи є D₁₀ абелевою групою?", G.is_abelian())

# Z₁₆ - циклічна група порядку 16
H = CyclicPermutationGroup(16)
print("Таблиця Келі для Z₁₆:")
print(H.cayley_table())
print("Порядок групи Z₁₆:", H.order())
print("Чи є Z₁₆ абелевою групою?", H.is_abelian())

# A₅ - альтернована група порядку 5
K = AlternatingGroup(5)
print("Таблиця Келі для A₅:")
print(K.cayley_table())
print("Порядок групи A₅:", K.order())
print("Чи є A₅ абелевою групою?", K.is_abelian())

# SL(2, Z₃) - група матриць 2x2 з визначником 1 над полем Z₃
L = SL(2, GF(3))
print("Таблиця Келі для SL(2, Z₃):")
print(L.cayley_table())
print("Порядок групи SL(2, Z₃):", L.order())
print("Чи є SL(2, Z₃) абелевою групою?", L.is_abelian())

# D₁₂ - діедральна група порядку 12
M = DihedralGroup(6)
print("Порядок групи D₁₂:", M.order())
print("Чи ізоморфні SL(2, Z₃) і D₁₂?", L.is_isomorphic(M))
```

## Завдання 2. 

__Знайдіть к-ть елементів кожного можливого порядку в групах $S_{100}$ та $A_{100}$.__

-----
___Зауваження___: вивід буде великий, збережіть його в окремий текстовий файл

___Зауваження 2___: оцініть спершу к-ть елементів в групі, а потім ще раз подумайте чи варто тут писати повний перебір

```python
# Для S_100
from sage.combinat.partition import Partitions

lcm_cache = {}

def count_orders_via_partitions(n):
    order_count = {}
    
    # Перебираємо всі можливі розбиття числа n
    for partition in Partitions(n):
        partition_tuple = tuple(partition)
        
        if partition_tuple in lcm_cache:
            order = lcm_cache[partition_tuple]
        else:
            # Якщо немає в кеші, обчислюємо НСК і зберігаємо в кеш
            order = lcm(partition)
            lcm_cache[partition_tuple] = order
        
        if order in order_count:
            order_count[order] += 1
        else:
            order_count[order] = 1
    
    return order_count

S100_orders = count_orders_via_partitions(100)

with open("group_orders_output_S100.txt", "w") as f:
    f.write("Порядки елементів у S_100:\n")
    for order, count in sorted(S100_orders.items()):
        f.write(f"Порядок: {order}, Кількість: {count}\n")

print("Результати збережено у файл 'group_orders_output_S100.txt'")

# # Для А_100
# from sage.combinat.partition import Partitions

# # Кеш для запам'ятовування вже обчислених НСК
# lcm_cache = {}

# def is_even_permutation(partition):
#     # Перевіряємо, чи є перестановка парною (парна кількість циклів непарної довжини)
#     odd_cycles = sum(1 for p in partition if p % 2 == 1)
#     return odd_cycles % 2 == 0

# def count_orders_in_A100(n):
#     order_count = {}
    
#     # Перебираємо всі можливі розбиття числа n
#     for partition in Partitions(n):
#         partition_tuple = tuple(partition)
        
#         # Перевіряємо кеш для розбиття
#         if partition_tuple in lcm_cache:
#             order = lcm_cache[partition_tuple]
#         else:
#             # Якщо немає в кеші, обчислюємо НСК і зберігаємо в кеш
#             order = lcm(partition)
#             lcm_cache[partition_tuple] = order
        
#         if is_even_permutation(partition):
#             if order in order_count:
#                 order_count[order] += 1
#             else:
#                 order_count[order] = 1
    
#     return order_count

# A100_orders = count_orders_in_A100(100)

# with open("group_orders_output_A100.txt", "w") as f:
#     f.write("Порядки елементів у A_100:\n")
#     for order, count in sorted(A100_orders.items()):
#         f.write(f"Порядок: {order}, Кількість: {count}\n")

# print("Результати збережено у файл 'group_orders_output_A100.txt'")
```

## Завдання 3. 
___Для заданих натуральних $n, k$ ($1 <= n \le 1 000 000, 1 \le k \le n!$) визначте чи існує в групі $S_n$ елемент порядку $k$.___

----------

___Зауваження:___ зверніть увагу на межі, в яких задано $n, k$. 

```python
from sage.combinat.partition import Partitions

def is_valid_partition(partition, k):
    return lcm(partition) == k

def exists_element_of_order(n, k):
    if k > n:
        return False

    for partition in Partitions(n, max_part=k):
        if is_valid_partition(partition, k):
            return True

    return False

n = 100000
k = 990

result = exists_element_of_order(n, k)

if result:
    print(f"У групі S_{n} існує елемент порядку {k}")
else:
    print(f"У групі S_{n} НЕ існує елементів порядку {k}.")
```
