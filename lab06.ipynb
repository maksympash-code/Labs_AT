## Завдання 3 
 
Задайте кілька діедральних груп $D_n$ для різних $n$. Як варіант, переберіть n в межах $3 \le n \le 100$. Для кожної діедральної групи побудуйте __список порядків нормальних підгруп__ (методом `normal_subgroups`). Знайдіть закономірність і виведіть гіпотезу.

Спрогнозуйте скільки нормальних підгруп має група $D_{470448}$ __без виводу всіх за допомогою Sage__. _Опишіть_ всі нормальні підгрупи групи $D_{470448}$.  

```python
results = []

for n in range(3, 101):
    Dn = DihedralGroup(n)
    normal_subgroups = Dn.normal_subgroups()
    orders = [G.order() for G in normal_subgroups]
    results.append((n, orders))

print("Аналіз порядків нормальних підгруп для D_n:")
for n, orders in results:
    print(f"D_{n}: Порядки нормальних підгруп = {orders}")

    
print("\nГіпотеза: порядки усіх нормальних підгруп порядку n = (тривіальній n = 1), дільникам n, самому n та n * 2")
def predict_normal_subgroups(n=470448):
    print(f"\nПрогноз для D_{n}:")
    print("Нормальні підгрупи:")
    print("- Тривіальна підгрупа: порядок = 1")
    print(f"- Підгрупа обертань: порядок = {n}")
    print(f"- Вся група: порядок = {2 * n}")
    print(f"Усі інші порядки нормальних підгруп є дільниками числа {n}.")
    
predict_normal_subgroups()
```
