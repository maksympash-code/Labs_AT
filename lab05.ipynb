## Завдання 1

Розгляньте групу матриць за множенням  

$$G = \left\langle \left(\begin{array}{rrr}
1 & 0 & 1 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{array}\right),  \left(\begin{array}{rrr}
1 & 0 & 0 \\
0 & 1 & 1 \\
0 & 0 & 1
\end{array}\right)\right\rangle$$

Дайте відповідь на наступні питання: 

1. _Чи є група G скінченною?_
2. _Чи є група G комутативною?_
3. _Який вигляд мають елементи групи G?_
4. _Які порядки мають елементи групи G?_ 
5. _Чому ізоморфна група G?_


__Відповіді__: 
1. Ні
2. Ні 
3. Група G нескінченна, тому не можливо порахувати елементи 
4. Група G нескінченна, тому не можливо визначити порядки
5. Ні

```python
# Задаємо матриці, що породжують групу
M1 = Matrix([[1, 0, 1], [0, 1, 0], [0, 0, 1]])
M2 = Matrix([[1, 0, 0], [0, 0, 1], [0, 1, 1]])

# Створюємо групу, породжену цими матрицями
G = MatrixGroup([M1, M2])

# 1. Чи є група скінченною?
is_finite = G.is_finite()
print(f"1. Чи є група G скінченною? {is_finite}")

# 2. Чи є група комутативною?
is_abelian = G.is_abelian()
print(f"2. Чи є група G комутативною? {is_abelian}")

# 3. Ізоморфність групи
D8 = DihedralGroup(4)
try:
    is_isomorphic = G.is_isomorphic(D8)
    print(f"5. Чи ізоморфна група G до D_8? {is_isomorphic}")
except NotImplementedError:
    print("Перевірка ізоморфності не підтримується для цієї групи.")

# 4. Перевірка, чи група скінченна: якщо так, виведемо елементи
if is_finite:
    elements = G.list()
    print(f"3. Елементи групи G:")
    for elem in elements:
        print(elem)

    print("4. Порядки елементів групи G:")
    for elem in elements:
        order = elem.order()
        print(f"Елемент:\n{elem}\nПорядок: {order}\n")
else:
    print("Група G нескінченна, не можна перерахувати всі елементи.")
```

## Завдання 2 

Розгляньте групу матриць за множенням: 

$$G = \left\langle \left(\begin{array}{rr}
1 & -1 \\
-1 & 2
\end{array}\right), \left(\begin{array}{rr}
3 & 2 \\
2 & 1
\end{array}\right) \right\rangle$$

Дайте відповідь на наступні питання: 

1. _Чи є група G скінченною?_
2. _Чи є група G комутативною?_
3. _Який вигляд мають елементи групи G?_
4. _Які порядки мають елементи групи G?_ 
5. _Чому ізоморфна група G?_


__Відповіді__: 
1. Ні
2. Так 
3. Група G нескінченна, тому не можливо порахувати елементи
4. Група G нескінченна, тому не можливо визначити порядки
5. Ні

```python
M1 = Matrix([[1, -1], [-1, 2]])
M2 = Matrix([[3, 2], [2, 1]])

G = MatrixGroup([M1, M2])

# 1. Чи є група скінченною?
is_finite = G.is_finite()
print(f"1. Чи є група G скінченною? {is_finite}")

# 2. Чи є група комутативною?
is_abelian = G.is_abelian()
print(f"2. Чи є група G комутативною? {is_abelian}")

# 3. Виводимо всі елементи групи, якщо вона скінченна
if is_finite:
    elements = G.list()
    print(f"3. Елементи групи G:")
    for elem in elements:
        print(elem)

    # 4. Виводимо порядки всіх елементів групи
    print("4. Порядки елементів групи G:")
    for elem in elements:
        order = elem.order()
        print(f"Елемент:\n{elem}\nПорядок: {order}\n")
else:
    print("Група нескінченна, тому неможливо перерахувати всі елементи.")

# 5. Перевіряємо ізоморфність групи з D_8 (діедральною групою порядку 8)
D8 = DihedralGroup(4)  # D_8 має порядок 8
try:
    is_isomorphic = G.is_isomorphic(D8)
    print(f"5. Чи ізоморфна група G до D_8? {is_isomorphic}")
except NotImplementedError:
    print("Перевірка ізоморфності не підтримується для цієї групи.")
```

$$K = \left\langle \left(\begin{array}{rrr}
-1 & 0 & 0 \\
0 & 1 & \frac{1}{2} \\
0 & 0 & 1
\end{array}\right), 
\left(\begin{array}{rrr}

0 & 0 & 1
\end{array}\right) \right\rangle$$

Дайте відповідь на наступні питання: 

1. _Чи є група K скінченною?_
2. _Чи є група K комутативною?_
3. _Який вигляд мають елементи групи K?_
4. _Які порядки мають елементи групи K?_ 

__Відповіді__: 
1. Ні
2. Ні 
3. Група K нескінченна, неможливо перерахувати всі елементи.
4. Група G нескінченна, тому не можливо визначити порядки.
5. Група K нескінченна, не можна обчислити всі підгрупи.


__Додаткова задача:__
Опишіть підгрупи групи K з точністю до ізоморфізму. 

```python
# Задаємо матриці, що породжують групу K
M1 = Matrix(QQ, [[-1, 0, 0], [0, 1, 1/2], [0, 0, 1]])
M2 = Matrix(QQ, [[1, 0, 1], [0, 1, 0], [0, 0, 1]])

# Створюємо групу, породжену цими матрицями
K = MatrixGroup([M1, M2])

# 1. Чи є група K скінченною?
is_finite = K.is_finite()
print(f"1. Чи є група K скінченною? {is_finite}")

# 2. Чи є група K комутативною?
is_abelian = K.is_abelian()
print(f"2. Чи є група K комутативною? {is_abelian}")

# 3. Вигляд елементів групи, якщо вона скінченна
if is_finite:
    elements = K.list()
    print(f"3. Елементи групи K:")
    for elem in elements:
        print(elem)

    print("4. Порядки елементів групи K:")
    for elem in elements:
        order = elem.order()
        print(f"Елемент:\n{elem}\nПорядок: {order}\n")
else:
    print("Група K нескінченна, неможливо перерахувати всі елементи.")

# Додаткова задача: Підгрупи групи K (обчислимо через породжені елементи)
def generate_subgroup(el):
    subgroup = set(el)
    size_b = 0
    while len(subgroup) > size_b:
        size_b = len(subgroup)
        new_el = {a * b for a in subgroup for b in subgroup}
        subgroup.update(new_el)
    return list(subgroup)

# Якщо група скінченна, обчислимо всі підгрупи
if is_finite:
    el = K.list()
    subgroup = generate_subgroup(el)
    print(f"Знайдено підгрупу з {len(subgroup)} елементів.")
    for el in subgroup:
        print(el)
else:
    print("Група K нескінченна, не можна обчислити всі підгрупи.")
```
