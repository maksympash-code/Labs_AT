## Завдання 1

Діедральна група $D_{20}$ має ряд нормальних підгруп: 

```python
G = DihedralGroup(20)
normal_subgroups = G.normal_subgroups()

for N in normal_subgroups:
    gens = G.gens()
    try:
        hom = PermutationGroupMorphism(G, G, [N(g) for g in gens])
        if hom.kernel() == N:
            print(f"Ендоморфізм для підгрупи порядку {N.order()} побудовано успішно.")
            print(f"Ядро ендоморфізму: {hom.kernel()}")
        else:
            print(f"Ядро не збігається з підгрупою порядку {N.order()}.")
    except ValueError as e:
        print(f"Не вдалося створити ендоморфізм для підгрупи порядку {N.order()}: {e}")
```

Для кожної нормальної підгрупи $N$ побудуйте ендоморфізм $\phi : D_{20} \rightarrow D_{20}$ такий, що $Ker \phi = N$. Задайте його через `PermutationGroupMorphism` і використайте метод `.kernel` як підтвердження. 

___Зауваження___: ядро має бути не __ізоморфне__ підгрупі, а __рівне__ їй.

## Завдання 2

Знайдіть центр та представників класів спряженості для груп $C_{11}, D_{13}, Q_{8}, SL(3, \mathbb{Z}_7)$. Виведіть кількість елементів в кожній групі, кількість класів спряженості та запишіть формулу для них. 

```python
# Групи
groups = {
    "C11": CyclicPermutationGroup(11),
    "D13": DihedralGroup(13),
    "Q8": QuaternionGroup(),
    "SL(3, Z7)": SL(3, GF(7))
}

# Виведення інформації для кожної групи
for name, G in groups.items():
    center = G.center()
    conjugacy_classes = G.conjugacy_classes()
    num_elements = G.order()
    num_classes = len(conjugacy_classes)
    
    print(f"\nГрупа {name}:")
    print(f"Кількість елементів: {num_elements}")
    print(f"Кількість класів спряженості: {num_classes}")
    print(f"Центр групи: {center}")
    print("Представники класів спряженості:", [c.representative() for c in conjugacy_classes])
```
